```python
import pandas as pd

def depurar_datos(archivo_2016, archivo_2023, salida):
    # Columnas necesarias
    columnas = ['Pais','Region','PIB per capita','Poblacion (millones)',
                'Huella ecologica total (consumo)','Biocapacidad total',
                'Deficit de biocapacidad','Tierras requeridas']

    # Cargar archivos
    df16 = pd.read_excel(archivo_2016)
    df23 = pd.read_excel(archivo_2023)

    # Normalizar nombres de países
    df16['Pais'] = df16['Pais'].astype(str).str.strip().str.lower()
    df23['Pais'] = df23['Pais'].astype(str).str.strip().str.lower()

    # Seleccionar columnas y eliminar filas con datos faltantes
    df16 = df16[columnas].dropna()
    df23 = df23[columnas].dropna()

    # Mantener solo países presentes en ambos archivos
    paises_comunes = set(df16['Pais']).intersection(set(df23['Pais']))
    df16 = df16[df16['Pais'].isin(paises_comunes)]
    df23 = df23[df23['Pais'].isin(paises_comunes)]

    # Agregar columna de año
    df16['Año'] = 2016
    df23['Año'] = 2023

    # Concatenar y guardar
    df_final = pd.concat([df16, df23], ignore_index=True)
    df_final.to_excel(salida, index=False)
    print("Archivo depurado:", salida)

    return df_final


#  GUARDAR el resultado de la función
df_final = depurar_datos(
    archivo_2016=r"C:/Users/HP/Documents/Jupiter/articulo, huella ecologica/huella ecologica 2016.xlsx",
    archivo_2023=r"C:/Users/HP/Documents/Jupiter/articulo, huella ecologica/Huella ecológica 2023.xlsx",
    salida=r"C:/Users/HP/Desktop/Jupiter/Datos_Depurados2.xlsx"
)

#  VERIFICACIÓN
print("Países únicos:", df_final['Pais'].nunique())
print("Filas por año:")
print(df_final['Año'].value_counts())
```
